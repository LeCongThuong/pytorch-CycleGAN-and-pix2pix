```python
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
def view_image(image_list_file_path):
    with open(image_list_file_path, 'r') as f:
        lines = f.read().splitlines()
    print("Done reading file")
    # num_images = len(lines)
    num_images = 81
    num_colums = 9
    # num_rows = num_images // num_colums + 1
    num_rows = 9
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_colums, figsize=(20, 16))
    print("Start render")
    for i in range(num_images):
        image_path = lines[i]
        pil_image = Image.open(image_path)
        axes[i // num_colums, i % num_colums].imshow(pil_image)
    plt.savefig("/home/love_you/Documents/Study/deep_learning/mocban/pytorch-CycleGAN-and-pix2pix/label_data_tools/small_area_image.jpeg")
    plt.show()
        
        
```

```python
image_list_file_path = '/home/love_you/Documents/Study/deep_learning/mocban/pytorch-CycleGAN-and-pix2pix/label_data_tools/small_area_image_paths_file.txt'

view_image(image_list_file_path)
```
